```python
# 연습1-1
import numpy as np
print('연습')

arr = np.loadtxt("./data/seoul.csv", delimiter=',', skiprows=1, 
                 dtype=np.object)
#print(arr)

for row in arr:
    if row[4] == '':
        row[4] = -999
    if row[3] == '':
        row[3] = 999
    if row[2] == '':
        row[2] = -999
    row[4] = float(row[4])
    row[3] = float(row[3])
    row[2] = float(row[2])
print(arr)

max_temp = 0
max_temp_date = ''
for i in arr:
    if i[4] > max_temp:
        max_temp_date = i[0]
        max_temp = i[4]
print('서울 최고로 더웠던 날 : ', max_temp_date)
print('최고기온 : ', max_temp)
    
```

```python
# 연습1-2
print(arr)

ilgyocha_max_temp = 0
ilgyocha_temp_date = ''
ilgyocha = 1
for i in arr:
    ilgyocha = i[4]-i[3]
    if ilgyocha > ilgyocha_max_temp:
        ilgyocha_temp_date = i[0]
        ilgyocha_max_temp = ilgyocha

print('서울 최고로 일교차 큰 날 : ', ilgyocha_temp_date)
print('최고 일교차 : ', ilgyocha_max_temp)
```

```python
# 연습2-1
arr_movie_rating = np.loadtxt("./data/ratings.csv", delimiter=',', 
                              skiprows=1, dtype=np.object)
for row in arr_movie_rating:
    row[0] = int(row[0])
    row[3] = float(row[3])
    row[2] = float(row[2])
    row[1] = float(row[1])
print(arr_movie_rating)

t_array = arr_movie_rating.T
print('전체 평점 평균 : ', np.mean(t_array[2]))  # 평균
```

```python
# 연습2-2
list_userid = []
for j in arr_movie_rating:
    list_userid.append(j[0])
# print(list_userid)
list_userid = set(list_userid)
print(list_userid)
print(len(list_userid))

countlist = [0]*610
sumlist = [0]*610
print(len(sumlist))
# print(countlist)
for k in list_userid:
    for m in arr_movie_rating:
        if k == m[0]:
            sumlist[k-1] = sumlist[k-1] + m[2]
            countlist[k-1] = countlist[k-1] + 1
    sumlist[k-1] = sumlist[k-1] / countlist[k-1]
# print(sumlist)  # 유저별 평균평점
for k in list_userid:
    print('유저 넘버 : ', k, ', 얘가 남긴 평균평점 : ', sumlist[k-1])
```
