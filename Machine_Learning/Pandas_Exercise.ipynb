```python
# 1-1
import numpy as np
import pandas as pd
df_m_rat = pd.read_csv("./data/ratings.csv")
df_movie = pd.read_csv("./data/movies.csv")

# print(df_m_rat.shape)
# print(df_movie.shape)
print(df_m_rat['rating'].mean())  # 전체 평균 평점
```

```python
# 1-2
# print(df_m_rat)
df_movie_rating_user = df_m_rat['rating'].groupby(df_m_rat['userId'])
display(df_movie_rating_user.mean())

# 유저별 평균 평점
```

```python
# 1-3
# print(df_m_rat)
df_movie_rating_movie = df_m_rat['rating'].groupby(df_m_rat['movieId'])
display(df_movie_rating_movie.mean())

# 영화별 평균 평점
```

```python
# 1-4
df_movie_rating_movie = df_m_rat['rating'].groupby(df_m_rat['movieId'])
# display(df_movie_rating_movie.mean()) # 영화별 평균평점

maxval = df_movie_rating_movie.mean().max() # 최대 평균평점 

# 합체
result = pd.merge(df_movie, df_movie_rating_movie.mean(), on='movieId', how='outer')

display(result.loc[result["rating"] == maxval, : ].sort_values(by='title'))
# 평균평점이 가장 높은 영화들
```

```python
# 1-5

# comedy 영화중 가장 평점이 낮은 영화 제목
# display(result)
new_result = result.dropna(how='any', inplace=False)
# display(new_result)
new_result_comedy = new_result.loc[new_result['genres'].str.contains('Comedy'), : ]
# str.contain()

minval = new_result_comedy['rating'].min() # 최소값

display(new_result_comedy.loc[new_result_comedy["rating"] == minval, : ].sort_values(by='title'))
```

```python
# 1-6

# 2015년에 평가된 모든 Romance영화의 평점은?

# print(df_m_rat)
# 1420038000 ~ 1451573999
new_df = df_m_rat.loc[df_m_rat["timestamp"] > 1420038000, : ]
rating2015 = new_df.loc[new_df["timestamp"] < 1451573999, : ]

rating_movie2015 = rating2015['rating'].groupby(rating2015['movieId'])
display(rating_movie2015.mean()) # 영화별 평균평점

result2015 = pd.merge(df_movie, rating_movie2015.mean(), on='movieId', how='outer')
new_result2015 = result2015.dropna(how='any', inplace=False)

# display(new_result)
new_result_romance = new_result2015.loc[new_result2015['genres'].str.contains('Romance'), : ]
display(new_result_romance)

romance_mean2015 = new_result_romance['rating'].mean()
print(romance_mean2015)
```

```python
# 3
import pandas
import savReaderWriter

with savReaderWriter.SavReader('./data/Koweps_hpc10_2015_beta1.sav', ioUtf8 = True) as reader:
    df = pandas.DataFrame(reader.all(), columns = [s for s in reader.header])

print(df.shape)      # # (16664, 957)
df['h10_g3'].head()  # 성별
df['h10_g4'].head()  # 태어난 연도
```
